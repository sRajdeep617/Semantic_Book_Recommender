```python
from transformers import pipeline
classifier = pipeline('text-classification',
                     model="j-hartmann/emotion-english-distilroberta-base",
                     top_k = None,
                     device = 0)
```

```python
emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_score = {label: [] for label in emotion_labels}

def calculate_max_emotion_scores(predictions):
  per_sentence_emotion_scores = {label: [] for label in emotion_labels}

  for prediction in predictions:
    sorted_prediction = sorted(prediction, key = lambda x: x["label"])

    for index, label in enumerate(emotion_labels):
      per_sentence_emotion_scores[label].append(sorted_prediction[index]["score"])

  return {label: np.max(score) for label, score in per_sentence_emotion_scores.items()}
```

```python
for i in range(10):
  isbn.append(books_df["isbn13"][i])
  sentences = books_df["description"][i].split(".")
  predictions = classifier(sentences)
  max_emotion_scores = calculate_max_emotion_scores(predictions)

  for label, score in max_emotion_scores.items():
    emotion_score[label].append(max_emotion_scores[label])
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}

for i in tqdm(range(len(books_df))):
    isbn.append(books_df["isbn13"][i])
    sentences = books_df["description"][i].split(".")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
emotions_df = pd.DataFrame(emotion_scores)
emotions_df["isbn13"] = isbn
```

```python
emotions_df.head()
```

```python
books_df = pd.merge(books_df, emotions_df, on = "isbn13")
books_df.head()
```

```python
books_df.to_csv("books_with_emotions.csv", index = False)
```
